## Tanh Vs. Linear Output Analysis

This notebook illustrates early performance gains in a model using a tanh activation on the output layer agaubst a model using only a linear output.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Read data from simulation output into file

Game number was not explicitly included in the simulation function, we we're going to add it on with a simple list comprehension.

```python
data = pd.read_csv('output_comparison.txt')
data['Game Number'] = np.hstack([[500 * i] * 6 for i in range(1, 21)])
```

```python
grouped_learner = data[data['Test Agent'] == 'Random'].groupby('Learning Agent')
```

## Hyperbolic Tangent on Output Layer - Performance

Since the number of games was small (10000) we're going to only graph progress against the random agent. 
Below, we can see the rapid rise in the win rate as time goes on.

```python
grouped_learner.get_group('tanh output').plot(x='Game Number')
plt.ylabel('Ratio')
plt.title('Performance over 10 thousand games with TANH activation')
```

## Linear Output Performance

The linear output did far poorer than the non-linear output did. It's not clear that any performance gain even occurred over the course of these games.

```python
grouped_learner.get_group('linear output').plot(x='Game Number')
plt.ylabel('Ratio')
plt.title('Performance over 10 thousand games with linear activation')
```
