```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('algorithm_comparison.txt')
```

```python
grouped = data.groupby('Learning Agent')
```

```python
grouped.get_group('basic q-learning algorithm').plot(x='Game')
plt.title('Progress of Basic Q-Learning Algorithm')
plt.ylabel('Ratio')
```

```python
grouped.get_group('modified q-learning algorithm').plot(x='Game')
plt.title('Progress of Modified Algorithm')
plt.ylabel('Ratio')
```
